# Chart Visualization

```python
import pandas as pd
import pandas as ps
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm, rc


font_fname = r'C:\\Windows\\Fonts\\malgun.ttf'
font_family = fm.FontProperties(fname=font_fname).get_name()
plt.rcParams['font.family']=font_family
```

```python
values = [100, 400, 900]
years = ['2018', '2019', '2020']

plt.figure(figsize=(5, 4))
plt.bar(years, values, width=0.3)
plt.xticks(years)
plt.show()
```

## pie chart

- 특정 열의 데이터 부각 시킬때 explode
- 데이터가 어떤 컬럼의 부분인지 표시하고 싶을때 labels에 컬럼명 지정
- 원 안에 데이터 값을 표시하고 싶을때 autopct에 포맷 스트링 문자열 입력
- 도형에 그림자 설정 shadow

```python
plt.pie(values,
        explode=(0, 0, 0.1), labels=years,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.show()
```

```python
df = ps.DataFrame({'column1':[55, 44, 66],
                   'column2':[88, 78, 99]})
df['column2'].plot.pie()
plt.legend(loc=1)  # 1~4, lower-left, ..
plt.show()
```

- Series data with pie

```python
np.random.rand(4) * 3
```

```python
series = pd.Series(np.random.rand(4) * 3,
                   index=['a', 'b', 'c', 'd'],
                   name='series')
series.plot.pie(figsize=(6, 6));
```

```python
series.plot.pie(
    labels=['AA', 'BB', 'CC', 'DD'],
    colors=['r', 'g', 'b', 'c'],
    autopct='%.2f%%',
    fontsize=20,
    figsize=(6, 6),
);
```

np.random.rand()는 random으로 나온 숫자 중에 0도 섞일 수 있기 때문에 아래 예제 pie에서 df의 인덱스값 4개가 다 안나올 수도 있다. np.ramdon.randint()를 쓰는 이유는 python float으로 값이 나오기 때문에 빈 값이 없다.

```python
3 * np.random.rand(4, 2)
# 3 * np.random.randint(4, size=(4, 2))
```

y축에 대응하는 컬럼을 가지고 있는 DataFrame은 subplots=True일 경우 pie plot을 하게 되면 데이터 라벨링을 위한 plot.legend 기능에 자동으로 인식된다. legend=False를 하면 숨길 수 있다. labels, colors 키워드로 라벨과 컬럼을 설정할 수 있다.

```python
df = pd.DataFrame(
    # 3 * np.random.randint(4, size=(4, 2)),
    3 * np.random.rand(4, 2),
    index=['a', 'b', 'c', 'd'],
           columns=['x', 'y']
                  )
df.plot.pie(subplots=True, figsize=(8, 4));
```
