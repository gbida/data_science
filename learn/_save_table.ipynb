```python
import bs4
import csv
import os
import pandas
import urllib
```

```python
html = urllib.request.urlopen('http://www.pythonscraping.com')
bs = bs4.BeautifulSoup(html, 'html.parser')
home_img = bs.find('img', {'class': 'pagelayer-img'})
img_src = home_img['src']
print(img_src)
# urllib.request.urlretrieve(img_src, 'logo.jpg')  # save img file
```

- 특정 속성, src tag만 찾고 싶을때

```python
downloadList = bs.findAll(src=True)
print('bs.findAll(src=True)')
print(downloadList)
print()

print('all url["src"] in bs')
for url in downloadList:
	print(url['src'])
```

```python
def getAbsUrl(source):
	if source.startswith('http://www.'):
		url = f'http://{source[11:]}'
	elif source.startswith('http://'):
		url = source
	elif source.startswith('www'):
		url = f'http://{source[4:]}'
	else:
		url = source
	return url


def getDwnldPath(baseUrl, absUrl, downloadDir):
	path = absUrl.replace('www', '')
	path = path.replace(baseUrl, '')
	path = downloadDir + path
	directory = os.path.dirname(path)
	if not os.path.exists(directory):
		os.makedirs(directory)
	return path


downloadDir = 'downloaded'
baseUrl = 'https://pythonscraping.com'

for download in downloadList:
	fileUrl = getAbsUrl(download['src'])
	if fileUrl is not None:
		print(fileUrl)
		urllib.request.urlretrieve(fileUrl, filename=getDwnldPath(baseUrl, fileUrl, downloadDir))
```

- csv.writer()
  - `mode='w+'`- write or read mode
  - `mode='wt'` - write text mode
  - `dialect='excel'`
  - `newline=''` - prevent to add empty line

```python
csvfile = open('test.csv', mode='w', encoding='utf-8')

try:
	f = csv.writer(csvfile)
	f.writerow(('column1', 'column2'))
	for i in range(10):
		f.writerow((i+1, i+2, (i+1)*2))
finally:
	csvfile.close()
```

- save a table in html page as a csv

```python
html = urllib.request.urlopen('http://en.wikipedia.org/wiki/Comparison_of_text_editors')
bs = bs4.BeautifulSoup(html, 'html.parser')

# table = bs.find('table', {'class': 'wikitable'})
table = bs.select_one('table.wikitable')
rows = table.select('tr')

csvfile = open('text editors.csv', mode='wt', encoding='utf-8')
f = csv.writer(csvfile)
try:
	for row in rows:
		csvRow = []
		for cell in row.findAll(['td', 'th']):
			csvRow.append(cell.get_text())
		f.writerow(csvRow)
finally:
	csvfile.close()

pandas.read_csv('text-editors.csv')
```
